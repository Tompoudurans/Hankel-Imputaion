```python
def loopar(y,trm,non,sigma):
    mu = 0
    for i in range(non):
        bas = random.gauss(mu, sigma)
        y.append(autor(y[i:i+len(trm)],trm) + bas)
    return y

def autor(x,trm):
    y = 0
    for i in range(len(trm)):
        try:
            y = y + x[i]*trm[i]
        except IndexError:
            return y
    return y
```

```python
def for_hmat(vec,hoz,forcat,rep=numpy.nan):
    vert = 1 + forcat + len(vec) - hoz
    col = []
    for i in range(vert):
        row = numpy.array([])
        for j in range(hoz):
            if (i+j) < len(vec):
                row = numpy.append(row, vec[i + j,])
            else:
                row = numpy.append(row, rep)
        col.append(row)
    return numpy.array(col)

def extend_data(k, x, n):
    U = []
    for i in range(n - k + 1):
        temp = numpy.array([])
        for j in range(k):
            temp = numpy.append(temp, x[i + j,])
        U.append(temp)
    return U


def getpc(data,pc=1):
    compo = PCA(n_components=pc)#[:pc]
    fited = compo.fit(data)
    return fited
```

```python
def opt1(time,time2):
    timmat = scipy.linalg.hankel(time)
    tim2mat = scipy.linalg.hankel(time2)
    etvec,qtmat = numpy.linalg.eig(timmat)
    xmats = m(tim2mat,qtmat)
    totra = etvec - xmats
    return numpy.trace(totra)
```

```python
from scipy.spatial.distance import minkowski
from scipy.stats import ks_2samp, anderson_ksamp
```

```python
def dists(data):  # compute ICD
    num = data.shape[0]
    data = data.reshape((num, -1))
    dist = []
    for i in range(0, num - 1):
        for j in range(i + 1, num):
            dist.append(minkowski(data[i], data[j]))

    return numpy.array(dist)


def dist_btw(a, b):  # compute BCD
    a = a.reshape((a.shape[0], -1))
    b = b.reshape((b.shape[0], -1))
    dist = []
    for i in range(a.shape[0]):
        for j in range(b.shape[0]):
            dist.append(minkowski(a[i], b[j]))

    return numpy.array(dist)


def LS(real, gen):  # KS distance btw ICD and BCD
    dist_real = dists(real)
    dist_gen = dists(gen)  # ICD 2
    distbtw = dist_btw(real, gen)  # BCD

    D_Sep_1, _ = ks_2samp(dist_real, distbtw)
    D_Sep_2, _ = ks_2samp(dist_gen, distbtw)
    return 1 - numpy.max([D_Sep_1, D_Sep_2])  # LS=1-DSI
```

```python
def opt11(time1,time2):
    timmat = scipy.linalg.hankel(time1)
    tim2mat = scipy.linalg.hankel(time2)
    u1,s1,v1= scipy.linalg.svd(timmat)
    u2,s2,v2= scipy.linalg.svd(tim2mat)
    tot = 0
    for i in range(len(s1)):
        x1 =  s1[i]*u2[i]
        x2 =  s2[i]*u1[i]
        tot = tot + scipy.spatial.distance.sqeuclidean(x1,x2)
    return tot
```

```python
def testwithar(pram,amount,lengh,var):
    data = []
    y = [0.1]*len(pram)
    time = loopar(y,pram,lengh,var)
    for i in tqdm.tqdm(range(amount)):
        extra = loopar(y,pram,lengh,var*i)
        data.append(opt11(time,extra))
    return data
```
